```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
N=6
tvshow = (175,349,411,592,595,505)
movie = (251,836,1237,1424,1284,993)
index = np.arange(N)
plt.bar(x=index, height=tvshow, color='lightskyblue');
plt.bar(x=index, height=movie, color='plum', bottom=tvshow);
plt.xticks(ticks=index,labels=['2016','2017','2018','2019','2020','2021']);
plt.xlabel('amount of type')
plt.ylabel('release year')
plt.legend(['TV Show','Movie'])
plt.show()
```

```python
y = ['Phillipines','Egypt','Argentina','Thailand','Columbia','Brazil','Turkey','Not given','China','Mexico','Australia','Spain','France','Taiwan','India','Canada','Japan','South Korea','Pakistan','United states']
x = [3,14,20,24,25,27,30,30,41,48,53,53,65,71,81,82,169,172,350,1097]
plt.barh(y, x, color='lightskyblue')
plt.ylabel("Countries")
plt.xlabel("TV Show")
plt.title("TV Show Exporting")
plt.show()
```

```python
dicts = {"year":['2014','2015','2016','2017','2018','2019','2020','2021'],
            "movie":[24,82,426,1185,1648,2016,1879,1498],
            "tvshow":[19,56,251,836,1237,1424,1284,993]}
Cards = pd.DataFrame(dicts)

plt.plot(Cards['year'],Cards[['movie']], color='plum')
plt.plot(Cards['year'],Cards[['tvshow']], color='lightskyblue')
plt.ylabel("Amount")
plt.xlabel("Year")
plt.title("Movie and TV Show added to Netflix")
plt.legend(['Movie','TV Show'])

plt.show()
```
